```python
import os
import zipfile

kaggle_dir = '/root/.kaggle'
kaggle_json_path = os.path.join(kaggle_dir, 'kaggle.json')

if not os.path.exists(kaggle_dir):
    os.makedirs(kaggle_dir)

if os.path.exists('./kaggle.json'):
    !cp ./kaggle.json {kaggle_dir}/
    !chmod 600 {kaggle_dir}/kaggle.json
    print(f"Kaggle API key moved to {kaggle_dir} and permissions set.")
else:
    print("Warning: kaggle.json not found in /content/. Please ensure you have uploaded your kaggle.json file.")
    print("Attempting download with existing Kaggle configuration...")

os.environ['KAGGLE_CONFIG_DIR'] = kaggle_dir

!kaggle datasets download nelyg8002000/commercial-aircraft-dataset

!mv ./commercial-aircraft-dataset.zip ./data/

zip_file_path = './data/commercial-aircraft-dataset.zip'

extract_dir = "./data/"

if not os.path.exists(extract_dir):
    os.makedirs(extract_dir)

with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall(extract_dir)

print(f"Dataset extracted to {extract_dir}")

os.remove(zip_file_path)
print(f"Removed downloaded zip file: {zip_file_path}")
```
